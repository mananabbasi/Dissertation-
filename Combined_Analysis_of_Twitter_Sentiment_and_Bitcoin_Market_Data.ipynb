```python
import pandas as pd

# Verify column names
print("Twitter_Sentiments columns:", Twitter_Sentiments.columns)
print("filtered_data columns:", filtered_data.columns)

# Convert 'date' column in Twitter_Sentiments to datetime
Twitter_Sentiments['date'] = pd.to_datetime(Twitter_Sentiments['date'])

# Convert 'Timestamp' column in filtered_data to datetime and rename it to 'date'
filtered_data['date'] = pd.to_datetime(filtered_data['Timestamp'], unit='s')

# Make the 'date' column in filtered_data timezone-naive
filtered_data['date'] = filtered_data['date'].dt.tz_localize(None)

# Sort both DataFrames by date (required for pd.merge_asof)
Twitter_Sentiments = Twitter_Sentiments.sort_values('date')
filtered_data = filtered_data.sort_values('date')

# Merge based on the nearest timestamp using pd.merge_asof
combined_data = pd.merge_asof(
    Twitter_Sentiments,  # Left DataFrame
    filtered_data,       # Right DataFrame
    on='date',           # Key column to merge on
    direction='nearest'  # Merge based on the nearest timestamp
)

# Inspect the combined DataFrame
print(combined_data.head())
print(combined_data.info())

# Handle missing data (if needed)
# Option 1: Drop rows with missing values
combined_data.dropna(inplace=True)

# Option 2: Fill missing values with a default value (e.g., 0)
# combined_data.fillna(0, inplace=True)

# Check the final shape of the combined dataset
print("Combined dataset shape:", combined_data.shape)
```

```python
time_frames = [
    {
        "name": "Time Frame 1",
        "start": "2021-02-05 10:52:04",
        "end": "2021-06-23 19:29:37"
    },
    {
        "name": "Time Frame 2",
        "start": "2022-04-07 13:00:00",
        "end": "2022-08-30 12:31:00"
    },
    {
        "name": "Time Frame 3",
        "start": "2023-02-25 20:49:09",
        "end": "2023-03-05 23:59:56"
    }
]
```

```python

# Ensure the 'date' column is in datetime format
combined_data_cleaned['date'] = pd.to_datetime(combined_data_cleaned['date'])

# Filter data for each time frame and store in a dictionary
filtered_data = {}
for tf in time_frames:
    # Convert start and end to datetime
    start_time = pd.to_datetime(tf['start'])
    end_time = pd.to_datetime(tf['end'])

    # Create a mask to filter rows within the time frame
    mask = (combined_data_cleaned['date'] >= start_time) & (combined_data_cleaned['date'] <= end_time)

    # Apply the mask and store the filtered data
    filtered_data[tf['name']] = combined_data_cleaned[mask]

# Inspect the filtered data for each time frame
for name, data in filtered_data.items():
    print(f"Data for {name}:")
    print(data.head())
    print(data.info())
    print("\n")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming filtered_data is a dictionary with time frames as keys
for tf_name, data in filtered_data.items():
    # Create a figure with three subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 18), sharex=True)

    # Plot Bitcoin closing price over time
    ax1.plot(data['date'], data['Close'], color='blue', label='Bitcoin Closing Price')
    ax1.set_ylabel('Price (USD)')
    ax1.set_title(f'{tf_name}: Bitcoin Closing Price Over Time')
    ax1.legend()
    ax1.grid(True, linestyle='--', alpha=0.7)

    # Plot compound sentiment over time
    ax2.plot(data['date'], data['compound'], color='green', label='Compound Sentiment')
    ax2.set_ylabel('Compound Sentiment')
    ax2.set_title(f'{tf_name}: Compound Sentiment Over Time')
    ax2.legend()
    ax2.grid(True, linestyle='--', alpha=0.7)

    # Plot sentiment distribution over time
    sentiment_counts = data.groupby(['date', 'sentiment']).size().unstack(fill_value=0)
    ax3.plot(sentiment_counts.index, sentiment_counts['positive'], color='green', label='Positive Sentiment')
    ax3.plot(sentiment_counts.index, sentiment_counts['neutral'], color='blue', label='Neutral Sentiment')
    ax3.plot(sentiment_counts.index, sentiment_counts['negative'], color='red', label='Negative Sentiment')
    ax3.set_xlabel('Date')
    ax3.set_ylabel('Sentiment Count')
    ax3.set_title(f'{tf_name}: Sentiment Distribution Over Time')
    ax3.legend()
    ax3.grid(True, linestyle='--', alpha=0.7)

    # Format x-axis dates
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Save the figure
    plt.savefig(f'{tf_name}_analysis.png', dpi=300, bbox_inches='tight')

    # Show the figure
    plt.show()
```

Doing Statistical Testing

T Test
Purpose: Compare the means of two groups (e.g., positive vs. negative sentiment) to see if they are significantly different.

Use Case: Test if the average Bitcoin price (Close) is significantly different during positive sentiment vs. negative sentiment.

```python
from scipy.stats import ttest_ind

# Split data into positive and negative sentiment groups
positive_sentiment = combined_data_cleaned[combined_data_cleaned['sentiment'] == 'positive']['Close']
negative_sentiment = combined_data_cleaned[combined_data_cleaned['sentiment'] == 'negative']['Close']

# Perform t-test
t_stat, p_value = ttest_ind(positive_sentiment, negative_sentiment)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

Reject the nukll hypothesis
If the p-value is below 0.05, it suggests that the average Bitcoin price is significantly different between positive and negative sentiment.

3. ANOVA for Sentiment and Price
Purpose: Compare the means of more than two groups (e.g., positive, neutral, negative sentiment).

Use Case: Test if the average Bitcoin price (Close) differs significantly across sentiment categories.

```python
from scipy.stats import f_oneway

# Split data into sentiment groups
positive_sentiment = combined_data_cleaned[combined_data_cleaned['sentiment'] == 'positive']['Close']
neutral_sentiment = combined_data_cleaned[combined_data_cleaned['sentiment'] == 'neutral']['Close']
negative_sentiment = combined_data_cleaned[combined_data_cleaned['sentiment'] == 'negative']['Close']

# Perform ANOVA
f_stat, p_value = f_oneway(positive_sentiment, neutral_sentiment, negative_sentiment)
print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

Since the p-value is very small, we reject the null hypothesis and conclude that there is a statistically significant difference in Bitcoin closing prices across the sentiment group

Chi-Square Test for Sentiment and Volume
Purpose: Test the independence of two categorical variables.

Use Case: Test if sentiment categories (positive, neutral, negative) are independent of trading volume categories (e.g., high, medium, low).

```python
from scipy.stats import chi2_contingency

# Create volume categories (e.g., high, medium, low)
combined_data_cleaned['volume_category'] = pd.cut(combined_data_cleaned['Volume'], bins=3, labels=['low', 'medium', 'high'])

# Create a contingency table
contingency_table = pd.crosstab(combined_data_cleaned['sentiment'], combined_data_cleaned['volume_category'])

# Perform Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2_stat}, P-value: {p_value}")
```

The p-value is greater than the common significance level of 0.05, indicating that there is no statistically significant association between sentiment and volume categories.

We fail to reject the null hypothesis, which states that sentiment and volume categories are independent.
